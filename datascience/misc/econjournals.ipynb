```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
journal_df = pd.read_csv('/Users/toddhendricks/Downloads/Journals.csv')
```

```python
journal_df.set_index(list(journal_df.columns)[0])
```

```python
journal_df.sort_values(by='subs',ascending=False)
```

```python
publisher_stats = journal_df.groupby(['publisher'])
```

```python
publisher_stats.size().sort_values(ascending=False)
```

```python
topic = journal_df.groupby(['field'])
```

```python
fields = topic.size().sort_values(ascending=False)[:5]
```
