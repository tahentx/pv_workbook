```python
example = soup + v_soup + m_soup + n_soup + ne_soup + d_soup + l_soup
```

```python
stop_words = ["yes", "I", "yea", "applause", "king", "we", "the", "of", "and", "to", "that", "it", "was", "in", "is", "you", "for", "that", "there", "this","be", "are", "jr", "as", "our", "they", "with", "on", "out", "not", "from", "but", "yeah", "have","all", "my", "will", "your", "us", "about", "for", "he", "at", "by", "him","who","were", "or", "do", "if", "going", "go", "sir", "would", "what", "their", "them", "thine", "so", "can", "must", "david", "luther", "an", "which", "had", "been", "some", "than", "an"]
```

```python
vectorizer = CountVectorizer(stop_words=stop_words)
doc_word = vectorizer.fit_transform(example)
doc_word.shape
```

```python
dtm_lsa = pd.DataFrame(doc_word.toarray(), index=example, columns=vectorizer.get_feature_names()).head(10)
dtm_lsa
```

```python
lsa = TruncatedSVD(n_components=7, n_iter=7, random_state=42)
doc_topic = lsa.fit_transform(doc_word)
lsa.explained_variance_ratio_
```

```python
lsa.explained_variance_ratio_.sum()
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python
display_topics(lsa, vectorizer.get_feature_names(), 4)
```

```python
vectorizer = CountVectorizer(stop_words=stop_words, ngram_range=(1,3))
matrix = vectorizer.fit_transform(example)
freqs = zip(vectorizer.get_feature_names(), matrix.sum(axis=0).tolist()[0])
print(sorted(freqs, key=lambda x: -x[1]))
```

```python
true_k = 7
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(matrix)
y_kmeans = model.predict(matrix)
```

```python
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    top_ten_words = [terms[ind] for ind in order_centroids[i, :7]]
    print("Cluster {}: {}".format(i, ' '.join(top_ten_words)))
```
