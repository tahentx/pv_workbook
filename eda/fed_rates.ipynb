```python
import requests
import pandas as pd
```

```python
url = 'https://www.federalreserve.gov/releases/h15/'
rates_source = requests.get(url)
rates_data = pd.read_html(rates_source.text)
rates_data = rates_data[0]
rates_data.head()
```

```python
rates_data.set_index('Instruments')
```
